```python
sns.countplot(netf['type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

```python
sns.countplot(netf['rating'])
sns.countplot(netf['rating']).set_xticklabels(sns.countplot(netf['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating')
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netf)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
labels = ['Movie', 'TV show']
size = netf['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

# **pie chart for rating**

```python
netf['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

# **word cloud**

# ***country***

```python
from wordcloud import WordCloud
```

```python
plt.subplots(figsize=(25,15))
wordcloud = WordCloud(
                          background_color='yellow',
                          width=1920,
                          height=1080
                         ).generate(" ".join(df.country))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('country.png')
plt.show()
```

# **cast with in the shows**

```python
plt.subplots(figsize=(25,15))
wordcloud = WordCloud(
                          background_color='black',
                          width=1920,
                          height=1080
                         ).generate(" ".join(df.cast))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('cast.png')
plt.show()
```
